```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris =pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',
                  header= None,
                  names = ['sepal_length','sepal_width', 'petal_length', 'petal_wid','species']) # [꽃밭침 길이, 꽃받침 폭, 꽃잎의 길이, 꽃잎의 폭, 종류]

iris
```

```python
plt.hist(iris['sepal_length'], bins=10, edgecolor = 'black')
plt.title('Length of sepal')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
setosa_iris = iris[(iris['species'] == 'Iris-setosa')]
setosa_iris = setosa_iris.reset_index(drop=True)
setosa_iris.head(5)
```

```python
setosa_sepal_length_mean = setosa_iris['sepal_length'].mean()
setosa_sepal_length_mean
```

```python
setosa_petal_length_mean = setosa_iris['petal_length'].mean()
setosa_petal_length_mean
```

```python
versicolor_iris = iris[(iris['species'] == 'Iris-versicolor')]
versicolor_iris.head()
```

```python
versicolor_sepal_length_mean = versicolor_iris['sepal_length'].mean()
versicolor_sepal_length_mean
```
