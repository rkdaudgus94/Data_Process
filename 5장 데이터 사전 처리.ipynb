```python
df_auto = pd.read_csv("D:/sample_data/auto-mpg.csv",header= None)

df_auto.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year', 'origin','name']
display(df_auto.head())
```

```python
df_auto.dropna(subset=['horsepower'], inplace=True)
```

```python
df_auto['origin'].replace({1:'USA', 2:'EU', 3:'JAPAN'}, inplace = True)
df_auto.head()
```

```python
df_auto['origin'].unique()
```

```python
describe_dic = df_auto['horsepower'].describe() # 구간 분할
describe_dic
```

```python
bin_dividers = [describe_dic.loc['min'], describe_dic.loc['50%'],
               describe_dic.loc['75%'],describe_dic.loc['max']]
bin_dividers
```

```python
bin_names =['저출력','보통출력','고출력']
```

```python
df_auto['hp_bin'] = pd.cut(x=df_auto['horsepower'],
                          bins=bin_dividers,
                          labels=bin_names,
                          include_lowest= True)
df_auto.head(30)
```

```python
horsepower_dummies = pd.get_dummies(df_auto['hp_bin']) # 0과 1인 2개의 값으로 표현
horsepower_dummies.head(30)
```

```python
ts_3m = pd.date_range('2019-01-01',periods=6,
                     freq='3M', # 3개월
                     tz='Asia/Seoul')
ts_3m
```
